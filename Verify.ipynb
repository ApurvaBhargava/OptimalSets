```python
import pandas as pd
import numpy as np
```

```python
dataset_names = ['eo']
embed_types = ['cvec_pca16', 'cvec_nmf16', 'cvec_umap16', 'cvec_tsne16', 'bert', 'roberta', 'distil', 'glove6B', 'universal']
selection_types = ['kmeans_0', 'kld', 'ks', 'cos', 'recon', 'taddy_0']
all_list= ['indices_eo_'+e+'_'+s+'.txt' if 'recon' not in s else 'indices_eo_'+e+'_'+s+'.npy' for e in embed_types for s in selection_types] + ['indices_eo_random_0.txt', 'indices_eo_topics_taddy_0.txt']
counts = [100, 200, 300, 400, 700, 1000, 1400, 1800, 2400, 3000, 3600, 4200]
```

```python
dct = {}
for file in all_list:
  if 'kmeans' in file or 'random' in file:
    with open(file, 'r') as fh:
      text = fh.read()
      text = eval(text.split('\n')[5])
  elif 'kld' in file or 'cos' in file or 'ks' in file:
    with open(file, 'r') as fh:
      text = fh.read()
      text = eval(text)[:1000]
  elif 'taddy' in file:
    with open(file, 'r') as fh:
      text = fh.read()
      text = eval(text)[5]
  elif 'recon' in file:
    with open(file, 'r') as fh:
      text = np.load(file, allow_pickle=True)[:1000]
  else:
    text = "THIS WAS MISSED"
  print(file)
  print(len(text))
  dct[file] = text
```

```python
all_list = [x[:-4] for x in all_list]
matrix = pd.DataFrame(data=None, index=all_list, columns=all_list)
```

```python
for key1, val1 in dct.items():
  for key2, val2 in dct.items():
    matrix.at[key1[:-4], key2[:-4]] = len(set(val1).intersection(set(val2)))
matrix
```

```python
matrix.to_csv("matrix_of_commons.csv")
```
