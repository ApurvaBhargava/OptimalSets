```python
time_dict = {}
embed_types = ['cvec_pca16', 'cvec_nmf16', 'cvec_umap16', 'cvec_tsne16', 'bert', 'roberta', 'distil', 'glove6B', 'universal']
samps = [100, 500, 2000, 4000]
for x in samps:
  for embed in embed_types:
    time_dict[(embed+'_kmeans', x)] = dcts[embed] + dcts[('kmeans_'+embed, x)]
    time_dict[(embed+'_kld', x)] = dcts[embed] + dcts['kld_'+embed] + dcts[('farthest_point_sampler_'+embed, x)]
    time_dict[(embed+'_ks', x)] = dcts[embed] + dcts['ks_'+embed] + dcts[('farthest_point_sampler_'+embed, x)]
    time_dict[(embed+'_cos', x)] = dcts[embed] + dcts['cos_'+embed] + dcts[('farthest_point_sampler_'+embed, x)]
    time_dict[(embed+'_recon', x)] = dcts[embed] + dcts['recon_'+embed]
    time_dict[(embed+'_taddy', x)] = dcts[embed] + dcts[('topic_mat_'+embed, x)]
  time_dict[('random_noembed', x)] = dcts[('random_noembed', x)]
  time_dict[('topics_taddy', x)] = dcts['r_topics'] + dcts[('topic_mat_taddy', x)]
```

```python
time_df = pd.DataFrame.from_dict(time_dict, orient='index', columns=['time in seconds'])
time_df.index = pd.MultiIndex.from_tuples(time_df.index, names=['method', 'samps'])
time_df.reset_index(level=['method', 'samps'], inplace=True)
time_df.pivot(index='samps', columns='method', values='time in seconds')
```

```python
time_df.pivot(index='samps', columns='method', values='time in seconds').to_csv("time_df.csv")
```
